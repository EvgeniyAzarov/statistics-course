```python
import numpy as np
from scipy import stats
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = (20, 6)
```

```python
import pandas as pd
```

```python
solves = pd.read_csv("/home/evgeniy/Загрузки/csTimerExport_20220604_143720.csv", index_col='No.', sep=';')
```

```python
solves
```

```python
plt.hist(solves['Time'], density=True, alpha=0.9, color='gray')
x = np.linspace(solves['Time'].min(), solves['Time'].max(), 100)
plt.plot(x, stats.norm.pdf(x, solves['Time'].mean(), solves['Time'].std()), lw=4)
plt.show()
```

```python
stats.ks_1samp(solves['Time'], stats.norm.cdf, (solves['Time'].mean(), solves['Time'].std()))
```
